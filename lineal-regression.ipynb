```python
import pandas
import numpy as np
import xlrd, xlwt
import random
import math

RANGE = 200

def get_excel_data():
    rb = xlrd.open_workbook("~/CNYRUB_TOM_2.xlsx")
    sheet = rb.sheet_by_index(0)
    o = sheet.col_values(1)[1:RANGE+1]
    h = sheet.col_values(2)[1:RANGE+1]
    l = sheet.col_values(3)[1:RANGE+1]
    c = sheet.col_values(4)[1:RANGE+1]
    t = sheet.col_values(6)[1:RANGE+1]
    return t, o, h, l, c

t, o, h, l, c = get_excel_data()
```

```python
def get_med(x):
    xmed = 0
    for i in range(RANGE):
        xmed = xmed + float(x[i])
    xmed = xmed / RANGE
    return xmed

def get_double_med(x, y):
    xymed = 0
    for i in range(RANGE):
        xymed = xymed + float(y[i]) * float(x[i])
    xymed = xymed / RANGE
    return xymed
    
def get_medsq(x):
    xmedsq = 0
    for i in range(RANGE):
        xmedsq = xmedsq + float(x[i]) * float(x[i])
    xmedsq = xmedsq / RANGE
    return xmedsq

def get_b(x, y):
    preresult = 0.0
    for i in range(RANGE):
        preresult+= float(x[i])*float(x[i]) # get (Xt * X)
    preresult = 1 / preresult # get (Xt * X)^-1
    result = [] # Xt * Y
    for i in range(RANGE):
        result.append(float(x[i])*float(y[i])*preresult) # (Xt * X)^-1 * Xt * Y
    return result   

def print_det_koefs(xArray,nameArray):
    for i in range(len(xArray)):
        x = xArray[i]
        xmed = get_med(x)
        cmed = get_med(c)
        cmedsq = get_medsq(c)
        xcmed = get_double_med(x, c)
        xmedsq = get_medsq(x)
        xb1 = get_b(x,c)[1]
        xb0 = get_b(x,c)[0]
        cx = get_newy(xb0, xb1, x)
        det_for_x = get_corr(xcmed, xmed, cmed, math.sqrt(xmedsq - xmed * xmed), math.sqrt(cmedsq - cmed * cmed)) * get_corr(xcmed, xmed, cmed, math.sqrt(xmedsq -xmed * xmed),math.sqrt(cmedsq - cmed * cmed))
        print("Ð¡oefficient of determination for {} = {}".format(nameArray[i],det_for_x))

def get_newy(b0, b1, x):
    y = []
    for i in range(RANGE):
        y.append(b0 + b1 * float(x[i]) + 2.7)
    return y

def get_sum_2(x):
    sum = 0
    for i in range(RANGE):
        sum = sum + float(x[i]) * float(x[i])
    return sum
```

```python
def get_corr(xymed, xmed, ymed, xsqrt, ysqrt):
    r = (xymed - xmed * ymed) / (xsqrt * ysqrt)
    return r
```

```python
print("TASK 1.1-1.4")
def get_c_of_x(X,e): # function of regression C(X) = B0+B1*X+e
    b = get_b(X,c)
    def f(Xi):
        return b[0]+b[1]*float(Xi)+e
    return f   

e = 0.01
parameters = [t,o,l,h]
parametersName = ["TIME","OPEN","LOW","HIGH"]
RANGE_PRINTED = 100
for j in range(len(parameters)):
    f = get_c_of_x(parameters[j],e)
    for i in range(min([RANGE_PRINTED,RANGE])):
        print("{} C({}) = {}").format(parametersName[j],parameters[j][i],f(parameters[j][i]))
```

```python
def get_b_M(xArray, yArray):
    m = []
    y = []
    for i in range(RANGE):
        newLine = [1]
        for x in xArray:
            newLine.append(float(x[i]))
        m.append(newLine)
        y.append([float(yArray[i])])
    mat = np.matrix(m)
    first = mat.T * mat
    second = first.I
    b = second * mat.T * y
    return b  # (Xt * X)^-1 * Xt * Y
def get_c_of_multy(xArray,e): # function of multi regression C(X1,X2,..., Xn) = B0 +B1*X1 + B2*X2 + BnXn + e
    b = get_b_M(xArray,c)
    def f(xiArray):
        result = b[0] + e
        for i in range(len(xArray)):
            result += float(b[i+1])*float(xiArray[i])
        return result
    return f        

print("TASK 2.1")

f_2_1 = get_c_of_multy([h,o],0.01)

for i in range(RANGE):
    print("C({},{}) = {}".format(h[i],o[i],f_2_1([h[i],o[i]])))

print("TASK 2.3")
f_2_3 = get_c_of_multy([h,l,o],0.01)

for i in range(RANGE):
    print("C({},{}) = {}".format(h[i],o[i],f_2_1([h[i],l[i],o[i]])))
   

```

```python
print("TASK 2.2")
def get_r2(x,y):
    xmed = get_med(x)
    ymed = get_med(y)
    xymed = get_double_med(x,y)
    dx = (get_sum_2(x)/len(x)) - xmed*xmed
    dy = (get_sum_2(y)/len(y)) - ymed*ymed
    corr = (xymed - xmed*ymed)/(math.sqrt(dx*dy))
    return corr*corr
print("R2")
print(get_r2(o,c))
print(get_r2(h,c))
print(get_r2(l,c))
print(get_r2(t,c)) 
```
